### 1. Original formalism: nested Loop strategy with empty table

```python
for i in range(2):
  for j in range(3):
    for k in range(2):
      result = "TRUE" if (i == j or i == k) else ""
      print(f"i={i}, j={j}, k={k}: {result}")
```

```python
for i in range(2):
  for j in range(3):
    for k in range(1):
      result = "TRUE" if (i == j or i == k) else ""
      print(f"i={i}, j={j}, k={k}: {result}")            
```

```python
for i in range(2):
  for j in range(3):
    for k in range(0):
      result = "TRUE" if (i == j or i == k) else ""
      print(f"i={i}, j={j}, k={k}: {result}")            
```

### 2. New more intuitive formulation

```python
R = [1, 2]
S = [1]
T = [3]

print("\n--- Nested loop iteration and condition check ---")
for r in R:
  for s in S:
    for t in T:
      result = "TRUE" if (r == s or r == t) else ""
      print(f"r={r}, s={s}, t={t}: {result}")          
```
